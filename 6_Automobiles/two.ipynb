```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('autos.csv')
data.head()
```

```python
data.isnull().sum()
```

+ Average Price per Make

```python
data['Ave_Price_Make'] = data.groupby('make')['price'].transform('mean')
```

```python
data.head(3)
```

```python
data['fuel_system'].unique()
```

```python
df_train = data.sample(frac = 0.72)
df_valid = data.drop(df_train.index)
```

+ Average Horsepower by Fuel_System

```python
df_train['Av_HP_FuelS'] = df_train.groupby('fuel_system')['horsepower'].transform('mean')
df_train.head(3)
```

```python
df_valid.head(3)
```

```python
df_valid = df_valid.merge(
    df_train[['fuel_system', 'Av_HP_FuelS']].drop_duplicates(),
    on = 'fuel_system',
    how = 'left'
)

df_valid[['fuel_system', 'Av_HP_FuelS']].head()
```
