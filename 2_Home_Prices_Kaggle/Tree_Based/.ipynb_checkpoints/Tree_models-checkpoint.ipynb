```python
import pandas as pd


from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.svm import SVR

from sklearn import metrics
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
num_data = data.select_dtypes(exclude='object')
num_data.head()
```

```python
num_data.isnull().sum()
```

```python
num_data = num_data.dropna(axis = 1)
num_data = num_data.drop('Id', axis=1)

num_data.isnull().sum()
```

```python
train = num_data.sample(frac=.75, random_state = 11)
valid = num_data.drop(train.index)

train.head(10)
```

```python
valid.head(10)
```

```python
X_train, y_train = train.drop('SalePrice', axis = 1), train.SalePrice
X_valid, y_valid = valid.drop('SalePrice', axis = 1), valid.SalePrice

X_train.head(3)
```

```python
model_dt = DecisionTreeRegressor(random_state=11)
model_dt.fit(X_train, y_train)
```

```python
preds_dt = model_dt.predict(X_valid)

mae = metrics.mean_absolute_error(y_valid, preds_dt)

print('DecisionTree MAE ')
mae
```

```python
model_rf = RandomForestRegressor(random_state=4)
model_rf.fit(X_train, y_train)
```

```python
preds_rf = model_rf.predict(X_valid)

mae = metrics.mean_absolute_error(y_valid, preds_rf)
print('RandomForest MAE')
mae
```

```python
model_svr = SVR()
model_svr.fit(X_train, y_train)
```

```python
preds_svr = model_svr.predict(X_valid)

mae = metrics.mean_absolute_error(y_valid, preds_svr)
print('SupportVector MAE')
mae
```
