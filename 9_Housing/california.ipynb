```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# First Time >> Clustering
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('housing.csv', index_col = 0)
data.head()
```

```python
# Latitude, Longitude n MedianIncome
X = data[['Latitude', 'Longitude', 'MedInc']].copy()
X.head()
```

## Create cluster feature

```python
kmeans = KMeans(n_clusters = 6) # Six centroids
```

```python
X['cluster'] = kmeans.fit_predict(X) # Creating the cluster feature
X.head()
```

```python
X['cluster'].unique()
```

```python
X['cluster'].dtype
```

```python
# Convert the type to be Categorical
X['cluster'] = X['cluster'].astype('category')
```

```python
X['cluster'].dtype
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x = X['Longitude'], y = X['Latitude'], hue  = X['cluster'])
```

```python
plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=16,
    titlepad=10,
)
```

```python
sns.relplot(x = 'Longitude', y = 'Latitude', hue = 'cluster', data = X, height = 6)
```

```python
# BoxPlots 
# SHows the DIstribution of Target within cluster
X['MedHouseVal'] = data['MedHouseVal']

sns.catplot(x = 'MedHouseVal', y = 'cluster', data = X, kind='boxen', height = 6)
```
